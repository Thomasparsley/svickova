```python
import time
import pathlib

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
```

## Cílová URL

```python
URL = "https://menza.upol.cz/webkredit2/Ordering/Menu?canteen=1&dateFrom=2021-12-06&dateTo=2021-12-06"
```

## Inicializace headless prohlížeče
Je potřeba stahnout chromedriver https://chromedriver.chromium.org

```python
options = Options()
options.headless = True
service = Service(pathlib.Path("/usr/local/bin/chromedriver"))
driver = webdriver.Chrome(service=service, options=options)
```

## Stažení HTML stránky
Sleep je podstatný aby se stránka stihla vykreslit

```python
driver.get(URL)
time.sleep(0.1)
print(driver.page_source)
```
